```python
df = pd.read_csv("/home/ubuntu/Machine-Learning-Algorithms/titanic_test.csv")
df.head(5)
```

```python
df = pd.read_csv("/home/ubuntu/Machine-Learning-Algorithms/titanic_test.csv",skiprows=1)
print("The first row is skipped which consisted of the real headers, thus the first row of data has taken the place of the header")
df
```

```python
df = pd.read_csv("/home/ubuntu/Machine-Learning-Algorithms/titanic_test.csv",header=None)
print("By default first row appears as the header but when we set header=None, the first row which in this case consists of header is here considered as data and appears in data section")
df
```

```python
df = pd.read_csv("/home/ubuntu/Machine-Learning-Algorithms/titanic_test.csv",header=None,names=[])
print("To set new values to the column header we use names parameter, in which the column names are supplied in the form of list")
df
```

```python
df = pd.read_csv("/home/ubuntu/Machine-Learning-Algorithms/titanic_test.csv",header=None,names=["v1","v2","v3","v4","v5","v6","v7","v8","v9","v10","v11"])
df
```

```python
df = pd.read_csv("/home/ubuntu/Machine-Learning-Algorithms/titanic_test.csv",skiprows=1,header=None,names=["v1","v2","v3","v4","v5","v6","v7","v8","v9","v10","v11"])
print("skiprows=1 parameter skips the first row that consists of the headers for the columns.")
df
```

```python
df = pd.read_csv("/home/ubuntu/Machine-Learning-Algorithms/titanic_test.csv",na_values=[0,"Q"])
print("navalues parameter is used to replace some values(values that represent absence of data) with NaN value")
df
```
